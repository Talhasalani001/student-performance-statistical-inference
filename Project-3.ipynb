```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('student-scores.csv')
```

### Q1: If math_score of males is greater than females.

#### Setting up the Null, and Alternative hypothesis:

Null Hypothesis $H_{o}$ : math_score of males and females are same.

Alternative Hypothesis $H_{alt}$ : math_score of males is greater than the math_score of females.

```python
males = df[df['gender']=='male']['math_score']
females = df[df['gender']=='female']['math_score']


t_stat, p_value = stats.ttest_ind(males, females, alternative='greater')
print("t-statistic:", t_stat, "p-value:", p_value)
```

#### Interpretation of Results
- **t-statistic**: A positive t-statistic (2.24) indicates that the sample mean math score of males is higher than that of females.
- **p-value**: A p-value of 0.0127 is less than the significance level of 0.05, meaning that the observed difference in math scores between males and females is statistically significant.

#### Conclusion
Since the **p-value is less than 0.05**, we **reject the null hypothesis**. This suggests that **males have significantly higher math scores than females** based on the data.

### Q2: If math_score of Non Doctors is greater than Doctors (Hint: Use career_aspirations to find Doctor and Non Doctors)

```python
df.head()
```

#### Setting up the Null, and Alternative hypothesis:

Null Hypothesis $H_{o}$ : The mean math scores of non-doctors are less than or equal to the mean math scores of doctors.

Alternative Hypothesis $H_{alt}$ : The mean math scores of non-doctors are greater than those of doctors.

```python
non_doctors = df[df['career_aspiration']!='Doctor']['math_score']
doctors = df[df['career_aspiration']=='Doctor']['math_score']

t_stat, p_value = stats.ttest_ind(non_doctors, doctors, alternative='greater')
print("t-statistic:", t_stat, "p-value:", p_value)
```

#### Interpretation of Results
- **t-statistic**: A negative t-statistic (-7.39) indicates that the sample mean math score of non-doctors is lower than the sample mean math score of doctors. 
- **p-value**: A p-value of approximately 1 means that the difference between the two groups is not statistically significant, and we cannot reject the null hypothesis.

#### Conclusion
Since the **p-value is much greater than 0.05**, we **fail to reject the null hypothesis**. This suggests that **there is no evidence to conclude that non-doctors have higher math scores than doctors**. In fact, the negative t-statistic suggests the oppositeâ€”that doctors may have higher math scores, though the difference is not statistically significant.

### Q3 If weekly_self_study_hours of males are greater than females.

```python
df.head()
```

#### Setting up the Null, and Alternative hypothesis:

Null Hypothesis $H_{o}$ : The mean weekly self-study hours of males are equal to or less than those of females.

Alternative Hypothesis $H_{alt}$ : The mean weekly self-study hours of males are greater than those of females.

```python
male_study_hours = df[df['gender'] == 'male']['weekly_self_study_hours']
female_study_hours = df[df['gender'] == 'female']['weekly_self_study_hours']


t_stat, p_value = stats.ttest_ind(male_study_hours, female_study_hours, alternative='greater')
print("t-statistic:", t_stat, "p-value:", p_value)
```

#### Interpretation of Results

- **t-statistic**: A positive t-statistic (3.46) indicates that the sample mean weekly self-study hours of males is higher than that of females.
- **p-value**: A p-value of 0.00028 is much smaller than the significance level of 0.05, meaning that the observed difference in weekly self-study hours between males and females is statistically significant.

#### Conclusion
Since the p-value is much less than 0.05, we reject the null hypothesis. This suggests that males spend significantly more time on self-study than females based on the data.

### Q4 If math_score of no part_time_job male students are greater than part_time_job male students.

```python
df.head()
```

Null Hypothesis $H_{o}$ : The mean math scores of males with no part-time job are less than or equal to those of males with part-time jobs.

Alternative Hypothesis $H_{alt}$ : The mean math scores of males with no part-time job are greater than those of males with part-time jobs.

```python
no_job_males = df[(df['gender'] == 'male') & (df['part_time_job'] == False)]['math_score']
job_males = df[(df['gender'] == 'male') & (df['part_time_job'] == True)]['math_score']

t_stat, p_value = stats.ttest_ind(no_job_males, job_males, alternative='greater')
print("t-statistic:", t_stat, "p-value:", p_value)
```

#### Interpretation of Results

- **t-statistic**: A high positive t-statistic (5.80) suggests that male students without part-time jobs have significantly higher math scores than those with part-time jobs.
- **p-value**: A p-value of 4.38e-09 is extremely small and far below the significance level of 0.05. This means the observed difference is statistically significant.

#### Conclusion
Since the p-value is much less than 0.05, we reject the null hypothesis. This indicates that male students without part-time jobs have significantly higher math scores than those with part-time jobs based on the data.

### Q9 Is there any Association between gender and career_aspiration.

Null Hypothesis $H_{o}$ : There is no association between gender and career_aspiration.

Alternative Hypothesis $H_{alt}$ : There is an association between gender and career_aspiration.

```python
contingency_table = pd.crosstab(df['gender'], df['career_aspiration'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-square:", chi2, "p-value:", p)
```

#### Interpretation of Results

Chi-square statistic: A large Chi-square value (155.39) indicates a significant deviation from the assumption that gender and career aspiration are independent.
p-value: The p-value is extremely small (6.72e-25), far below the significance level of 0.05. This means that the observed association between gender and career aspiration is statistically significant.

#### Conclusion
Since the p-value is far less than 0.05, we reject the null hypothesis. This suggests that there is a significant association between gender and career aspiration based on the data. Gender and career aspirations are not independent of each other.

### Q10 Is there any Association between gender and part_time_job.

Null Hypothesis $H_{o}$ : There is no association between gender and part-time job.

Alternative Hypothesis $H_{alt}$ : There is an association between gender and part-time job.

```python
contingency_table = pd.crosstab(df['gender'], df['part_time_job'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print("Chi-square:", chi2, "p-value:", p)
```

#### Interpretation of Results

p-value: A p-value of 0.3156 is greater than the significance level of 0.05, indicating that the observed association between gender and part-time job status is not statistically significant.

#### Conclusion
Since the p-value is greater than 0.05, we fail to reject the null hypothesis. This suggests that there is no significant association between gender and part-time job status based on the data. Gender and part-time job status appear to be independent.
